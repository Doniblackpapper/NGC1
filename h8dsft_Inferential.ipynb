```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("covid19.csv")
df.head()
```

```python
df.info()
```

```python
pd.to_datetime(df['Date'])
df['Day'] = pd.DatetimeIndex(df['Date']).day
df['Year'] = pd.DatetimeIndex(df['Date']).year
case = df[['Date','New Cases','Day','Year']] 
print('Average new case for 2020: {}'.format(np.round(case.loc[case['Year']==2020,'New Cases'].mean())))
print('Average new case up till now: {}'.format(np.round(case['New Cases'].mean())))
```

```python
#pembuktian bahwa H0: μ=149 ,  H1: μ > 149
t_stat,p_val = stats.ttest_1samp(case['New Cases'], 149)
print('P-value:',p_val/2) #The p-value divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat)

#karena p value lebih kecil dari 0.05 maka h null diterima
```
